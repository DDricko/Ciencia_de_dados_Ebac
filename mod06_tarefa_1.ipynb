#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
import pandas as pd
```

```python
sinasc = pd.read_csv('sinasc_RO_2019.csv')
sinasc.head()
```

:####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
import matplotlib.pyplot as plt
import seaborn as sns

variables_to_plot = ['APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE']

for var in variables_to_plot:
    plt.figure(figsize=(10, 6))
    sns.countplot(data=sinasc, x=var, order=sinasc[var].value_counts().index, palette='viridis')
    plt.title(f'Frequência de {var}')
    plt.xlabel(var)
    plt.ylabel('Frequência')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS



```python
import matplotlib.pyplot as plt
import seaborn as sns

apgar5_por_consultas = sinasc.groupby('CONSULTAS')['APGAR5'].mean()

plt.figure(figsize=(8, 6))
sns.barplot(x=apgar5_por_consultas.index, y=apgar5_por_consultas.values, palette='viridis')
plt.title('APGAR5 médio por número de consultas')
plt.xlabel('Número de Consultas')
plt.ylabel('APGAR5 Médio')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
import matplotlib.pyplot as plt
import seaborn as sns

variables_to_plot = ['SEXO', 'GESTACAO', 'ESTCIVMAE', 'CONSULTAS']

fig, axes = plt.subplots(2, 2, figsize=(15, 10))
axes = axes.flatten() # Flatten the 2x2 array of axes to easily iterate

for i, var in enumerate(variables_to_plot):
    sns.countplot(data=sinasc, x=var, ax=axes[i], palette='viridis')
    axes[i].set_title(f'Frequência de {var}')
    axes[i].set_xlabel(var)
    axes[i].set_ylabel('Frequência')
    axes[i].tick_params(axis='x', rotation=45) # Removed ha='right'

plt.tight_layout()
plt.show()
```
