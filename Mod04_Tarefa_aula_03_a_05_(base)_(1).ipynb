## Exercício 1: Vestibular

Considere que a os dados gerados na célula abaixo contêm o número de acertos de 100 alunos em um vestibular para um curso de exatas, divididas pelos respectivos assuntos. Considere que cada assunto possui um número de questões conforme a tabela abaixo:

| assunto | número de questões |
|:---:|:---:|
| Matemática | 24 |
| Português | 18 |
| Geografia | 8 |
| Inglês | 8 |
| História | 8 |
| Física | 12 |
| Química | 12 |

Usando os comandos de operações com DataFrames que você aprendeu na Aula 03, calcule:

1. (operações com escalar) Calcule o percentual de acerto dos alunos por assunto.  
2. (operações entre *DataFrames) Calcule o total de acertos de cada aluno.  
3. Calcule o porcentual geral de cada aluno.  
4. Suponha que a nota de corte para a segunda fase seja 45. Quantos alunos tiveram nota maior que 45?  

````python
import pandas as pd
import numpy as np

np.random.seed(42)
df_mat = pd.DataFrame(np.random.randint(24, size=(100, 1)), columns=['Qt_acertos'])

df_por = pd.DataFrame(np.random.randint(18, size=(100, 1)), columns=['Qt_acertos'])

df_geo = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_ing = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_his = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_fis = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])

df_qui = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])
````

````python
# 1) Seu código aqui
# Cria Tabela com o total de acertos dos alunos em cada materia
df_total_geral = pd.concat([df_mat, df_por, df_geo, df_ing, df_his, df_fis, df_qui], axis=1)
df_total_geral.columns = ['Matemática', 'Português', 'Geografia', 'Inglês', 'História', 'Física', 'Química']
df_total_geral
````

````python

## Cria tabela auxiliar com o total de perguntas para cada materia
qtd_perguntas_dic = {'Matemática': 24, 'Português': 18, 'Geografia': 8, 'Inglês': 8, 'História': 8, 'Física': 12, 'Química': 12}

qtd_perguntas_dic

qtd_perguntas_serie = pd.Series(qtd_perguntas_dic)
qtd_perguntas_serie
#
````

````python
#Calcula a porcentagem de acertos
df_total_geral_porcentagem = (df_total_geral / qtd_perguntas_serie) * 100
df_total_geral_porcentagem.round(2)
````

````python
# 2) Seu código aqui
# Calcula o valor total de acertos de cada aluno
nota_final_aluno = df_total_geral.sum(axis=1)
nota_final_aluno
````

````python
# 3) Seu código aqui
# Calcula o percentual geral do aluno
qtd_perguntas = qtd_perguntas_serie.sum()
qtd_perguntas

perc_geral_aluno = (nota_final_aluno / qtd_perguntas) * 100
perc_geral_aluno.round(2)
````

````python
# 4) Seu código aqui
alunos_aprovados = perc_geral_aluno.loc[perc_geral_aluno > 45].count()
alunos_aprovados
````

````python
# Exibindo Dataframe com os resultados obtidos
df_total_geral['Nota_Final'] = nota_final_aluno
df_total_geral['Perc_Geral'] = perc_geral_aluno
df_total_geral['Aprovado'] = perc_geral_aluno > 45
df_total_geral

````
